```python

# Define the file path (replace with actual path if needed)
file_path = "dir_names.txt"  # Ensure this file exists

# Read the file and store content in a list
with open(file_path, "r") as file:
    directory_name = [line.strip() for line in file]
# Print the list
print(directory_name)
```

```python
master_formatted_train_name = []
for folder in directory_name:
    formatted_train_names = []
    file_path = os.path.join(folder, "category_data_filtered.csv")  # Construct file path
    out_path_formatted_names = os.path.join(folder, "fomatted_names.txt")  # Construct file path
    out_path_formatted_links = os.path.join(folder, "fomatted_links.txt")  # Construct file path
    
    # Check if the file exists
    if os.path.exists(file_path):
        # Read the CSV file into a DataFrame
        df = pd.read_csv(file_path)
        
        # print(folder, " : ", df.shape)
        # df["Formatted Name"] = df["Train Name"].str.replace(" ", "-") + "-" + df["Number"]
        # df["Formatted Name"] = df["Train Name"].str.replace(" ", "-") + "-" + df["Number"].astype(str)
        
        
        df["Number"] = df["Number"].astype(str)

        # Process each row and format the train name
        # for _, row in df.iterrows():
        #     formatted_name = row["Train Name"].replace(" ", "-") + "-" + str(row["Number"])
        #     formatted_train_names.append(formatted_name)
        #     master_formatted_train_name.append(formatted_name)
        #     print(formatted_name)
        for _, row in df.iterrows():
            if len(row["Number"]) == 5:  # Ensure the train number has exactly 5 digits
                formatted_name = row["Train Name"].replace(" ", "-") + "-" + row["Number"]
                formatted_train_names.append(formatted_name)
                master_formatted_train_name.append(formatted_name)
                print(formatted_name)
            else:
                formatted_name = row["Train Name"].replace(" ", "-") + "-0" + row["Number"]
                formatted_train_names.append(formatted_name)
                master_formatted_train_name.append(formatted_name)
                print(formatted_name)
        
        # with open("formatted_names.txt", "w") as file:
        with open(out_path_formatted_names, "w") as file:
            for name in formatted_train_names:
                file.write(name + "\n")

        with open(out_path_formatted_links, "w") as file:
            for name in formatted_train_names:
                name = "-".join(word.capitalize() for word in name.split("-"))        
                file.write("https://etrain.info/train/" +name + "/history\n")
        

with open("master_formatted_names.txt", "w") as file:
    for name in master_formatted_train_name:
        file.write(name + "\n")

# with open(out_path_formatted_links, "w") as file:
with open("master_formatted_links.txt", "w") as file:
    for name in master_formatted_train_name:
        name = "-".join(word.capitalize() for word in name.split("-"))
        file.write("https://etrain.info/train/" +name + "/history\n")


                
```
